```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pylab
import scipy
from scipy.stats import skewnorm 
```

### Normal Distribution

```python
# Creating two normally distributed arrays for Target and Control Groups
tg_size = 100
cg_size = 100

np.random.seed(7)

tg = np.random.normal(loc=37, scale=3, size=tg_size)
cg = np.random.normal(loc=36.4, scale=3, size=cg_size)
```

```python
# Calculating mean values for Target and Control Groups
tg_mean = np.mean(tg)
cg_mean = np.mean(cg)

# Creating Distributions plot for our arrays
fig, ax = plt.subplots(figsize=(10,6))

sns.distplot(tg, bins=10, ax=ax, label='TG', color='b')
sns.distplot(cg, bins=10, ax=ax, label='CG', color='g')

ax.axvline(linewidth=2, linestyle='--', color='b', x=tg_mean)
ax.axvline(linewidth=2, linestyle='--', color='g', x=cg_mean)

plt.title('Temperature of insects')
plt.legend()

# Adding annotations
ax.annotate(str(round(tg_mean,0)), xy=(tg_mean, 0), xytext=(tg_mean, 0), arrowprops=dict(facecolor='b', shrink=0.05),)
ax.annotate(str(round(cg_mean,0)), xy=(cg_mean, 0), xytext=(cg_mean, 0), arrowprops=dict(facecolor='g', shrink=0.05),)

# Saving plot
fig.savefig('distplot_normal.png')

plt.show()
```

#### T-test

```python
t_test = scipy.stats.ttest_ind(tg, cg)
```

```python
t_test
```

```python
'''
thus ttest_ind calculates two-sided test and we have statistics > 0 
then we have to divide p-value by 2 in order to get one-sided p-value

'''
p_value = t_test.pvalue/2
```

```python
p_value*100
```

### Skewed Distribution

```python
tg_size = 100
cg_size = 100

tg_a = 20
cg_a = 10


scipy.random.seed(7)

tg = skewnorm.rvs(tg_a, size=tg_size)
cg = skewnorm.rvs(cg_a, size=cg_size)

tg_mean = np.mean(tg)
cg_mean = np.mean(cg)

fig, ax = plt.subplots(figsize=(10,6))

sns.distplot(tg, bins=10, ax=ax, label='TG', color='b')
sns.distplot(cg, bins=10, ax=ax, label='CG', color='g')

ax.axvline(linewidth=2, linestyle='--', color='b', x=tg_mean)
ax.axvline(linewidth=2, linestyle='--', color='g', x=cg_mean)

plt.legend()


# Saving plot
fig.savefig('distplot_skewed_1.png')
plt.show()
```
